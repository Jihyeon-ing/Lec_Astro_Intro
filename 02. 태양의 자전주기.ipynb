```python
import astropy.io.fits as fits
```

data: Proba2/SWAP 174 \AA solar EUV image  
via Helioviewer (https://helioviewer.ias.u-psud.fr/)

```python
file = fits.open('swap_lv1_20120606_000010.fits')
```

```python
print(file)
```

```python
data = file[0].data
```

```python
header = file[0].header
```

```python
print(header['TELESCOP'])
print(header['DATE-OBS'])
print(header['CDELT1'])
```

```python
import numpy as np
image = np.array(data)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='gray')
plt.show()
```

```python
# adjustment of brightness
image = image.clip(1, np.max(image))
image = np.log(image)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='gray', origin='lower') # set the origin to the lower left point (as Cartesian Coordinates)
plt.show()
```

```python
%matplotlib qt5
# show the image in the external window
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='gray', origin='lower')
plt.show()
```

```python
%matplotlib inline
# show the image in this notebook
plt.figure(figsize=(10, 10))
plt.imshow(image, cmap='gray', origin='lower')
plt.axis('off')
plt.show()
```
